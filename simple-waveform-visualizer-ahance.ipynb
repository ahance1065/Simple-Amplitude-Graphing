```python
#comments for future reference (all audio Datasets used is made by me, Austin Hance)
"""
This can be used to compare certain features of a song. For example, 
if you wanted to re-create the drums from scratch, you can download the drum audio file from an 
instrument extractor, then import the file here to compare the waveforms to your own drum samples.
(most DAWs/Editors have these waveforms).
"""

#just some libraries "¯\(ツ)/¯"
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn

#importing librosa (a python library built to work with audio and perform analysis)
import librosa
import librosa.display
import IPython.display as ipd
import warnings
warnings.filterwarnings('ignore')

#importing the audio file from kaggle datasets
import os
os.listdir('/kaggle/input')
y, sr = librosa.load('/kaggle/input/ahance-custom-audio-samples/planless nights 140 bpm f major AHANCE.wav')

#creating audio graph
print('y:', y, '\n')
print('y shape:', np.shape(y), '\n')
print('Sample Rate (KHz):', sr, '\n')

print('Check Len of Audio:', 661794/22050)

audio_file, _ = librosa.effects.trim(y)

print('Audio File:', audio_file, '\n')
print('Audio File shape:', np.shape(audio_file))

plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = audio_file, sr = sr, color = "#DE5959");
plt.title("Sound Waves in planless nights 140 bpm f major AHANCE", fontsize = 24);

#next audio File

import os
os.listdir('/kaggle/input')
y, sr = librosa.load('/kaggle/input/ahance-custom-audio-samples/jaded eyes 131 bpm f sharp minor AHANCE.wav')

print('y:', y, '\n')
print('y shape:', np.shape(y), '\n')
print('Sample Rate (KHz):', sr, '\n')

print('Check Len of Audio:', 661794/22050)

audio_file, _ = librosa.effects.trim(y)

print('Audio File:', audio_file, '\n')
print('Audio File shape:', np.shape(audio_file))

plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = audio_file, sr = sr, color = "#24DDE5");
plt.title("Sound Waves in jaded eyes 131 bpm f sharp minor AHANCE", fontsize = 24);

#next audio file

import os
os.listdir('/kaggle/input')
y, sr = librosa.load('/kaggle/input/ahance-custom-audio-samples/burial 124 bpm c sharp major AHANCE.wav')

print('y:', y, '\n')
print('y shape:', np.shape(y), '\n')
print('Sample Rate (KHz):', sr, '\n')

print('Check Len of Audio:', 661794/22050)

audio_file, _ = librosa.effects.trim(y)

print('Audio File:', audio_file, '\n')
print('Audio File shape:', np.shape(audio_file))

plt.figure(figsize = (16, 6))
librosa.display.waveshow(y = audio_file, sr = sr, color = "#8824E5");
plt.title("Sound Waves in burial 124 bpm c sharp major AHANCE", fontsize = 24);
```
